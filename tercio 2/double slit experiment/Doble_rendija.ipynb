# Proyecto doble rendija

## Historia

El experimento de la doble rendija, también conocido como el experimento de Young, es uno de los más famosos en la historia de la física y es fundamental para entender los conceptos de la mecánica cuántica y la naturaleza ondulatoria de la luz y las partículas.

Antes de este experimento, había un debate entre si la luz era una partícula o una onda. Isaac Newton apoyaba la teoría corpuscular de la luz, argumentando que la luz estaba compuesta por partículas, mientras que otros científicos, como Christiaan Huygens, sostenían que la luz era una onda.

En la década de 1920, Louis de Broglie propuso que no solo la luz, sino también las partículas, como los electrones, tienen propiedades tanto de onda como de partícula, lo que se conoce como dualidad onda-partícula. Esta idea fue probada cuando se realizó el experimento de la doble rendija con electrones. Sorprendentemente, incluso las partículas individuales (como electrones) exhibieron un patrón de interferencia al pasar por las rendijas, pero solo cuando no se las observaba directamente.

Este hecho introduce un elemento clave de la mecánica cuántica: la observación. Si se intenta medir por cuál rendija pasa el electrón, el patrón de interferencia desaparece y el comportamiento del electrón parece el de una partícula. Pero si no se mide, el electrón actúa como una onda, pasando por ambas rendijas a la vez y creando un patrón de interferencia.

## Simulación

Para este laboratorio se pidio realizar una simulacion en python sobre el experimento de la doble rendija, se realizo el siguiente codigo


```python
import numpy as np

import matplotlib.pyplot as plt

# Parámetros del experimento
num_rendijas = 2
num_detectores = 1000
amplitud_onda = 1
longitud_onda = 1
distancia_rendijas = 1
distancia_pantalla = 100

# Posiciones de las rendijas
posiciones_rendijas = np.linspace(-distancia_rendijas / 2, distancia_rendijas / 2, num_rendijas)

# Posiciones de los detectores en la pantalla
posiciones_detectores = np.linspace(-50, 50, num_detectores)

# Función para calcular la amplitud en un punto de la pantalla
def calcular_amplitud(x):
    amplitud_total = 0
    for rendija in posiciones_rendijas:
        distancia = np.sqrt((x - rendija) ** 2 + distancia_pantalla ** 2)
        fase = (2 * np.pi / longitud_onda) * distancia
        amplitud_total += amplitud_onda * np.cos(fase)
    return amplitud_total

# Calcular la intensidad en cada punto de la pantalla
intensidades = [calcular_amplitud(x) ** 2 for x in posiciones_detectores]

# Graficar los resultados
plt.plot(posiciones_detectores, intensidades)
plt.title('Experimento de la Doble Rendija')
plt.xlabel('Posición en la pantalla')
plt.ylabel('Intensidad')
plt.show()
```

El resultado anterior simula el comportamiento del sistema fisico de la doble rendija el cual tiene mayor intensidad en el centro de las dos rendijas separadas a una distancia d y se va a tenuando a medida de la posicion de la pantalla como se puede ver en la figura anterior. Por lo que se puede decir que los resultados obtenidos tuvieron una buena estimación.
